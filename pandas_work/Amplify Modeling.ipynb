```python
#import Spotify steams data

file_2 = "streams_no_genre_low_streams_removed.csv"
appended_df = pd.read_csv(file_2)
appended_df.head()
```

```python
#merge beachland and streams data

merged_prototype=pd.merge(appended_df,raw_data_df,how='left', on='INDEX')
# merged_prototype=merged_prototype.drop(columns={'INDEX_y','DATE_y'})
```

```python
#rename / clean columns

merged_prototype=merged_prototype.rename(columns={'INDEX_x':'INDEX','DATE_x':'DATE'})
merged_prototype.head()
```

```python
#deduplicate index values for initial exploration 

Deduplicated_df = merged_prototype.drop_duplicates(subset='INDEX', keep='first', inplace=False)
print(f"Length of deduplicated dateframe: {len(Deduplicated_df)}")

# drop missing rows without streams
Deduplicated_df=Deduplicated_df.dropna(subset=['streams30s'])
print(f"Length of deduplicated dateframe with missing values removed: {len(Deduplicated_df)}")
```

```python
#remove outlier ticket sales and streams
Deduplicated_df = Deduplicated_df.loc[(Deduplicated_df['streams30s']<10000000)]
Deduplicated_df = Deduplicated_df.loc[(Deduplicated_df['streams30s']>500)]
Deduplicated_df = Deduplicated_df.loc[(Deduplicated_df['TOTAL COUNT']>5)]

#filter shows to ballroom and tavern only
Deduplicated_df = Deduplicated_df.loc[(Deduplicated_df['ROOM']=="BALLROOM") | (Deduplicated_df['ROOM']=="TAVERN")]
 
#sort, length, print
Deduplicated_df=Deduplicated_df.sort_values('streams30s',ascending=False)
print(f"New deduplicated dataframe length: {len(Deduplicated_df)}")
Deduplicated_df.head(10)
```

```python
Deduplicated_df['streams_transformed'] = Deduplicated_df['streams30s']/8786298
Deduplicated_df.head()
```

```python
#create day of week and group by for means
Deduplicated_df['DATE'] = pd.to_datetime(Deduplicated_df['DATE'])
Deduplicated_df['Day of Week'] = Deduplicated_df['DATE'].dt.day_name()
Deduplicated_df.groupby('Day of Week').mean()
```

```python
Deduplicated_df['Month'] = Deduplicated_df['DATE'].dt.month
month_analysis = Deduplicated_df.groupby('Month').mean()
month_analysis = month_analysis.loc[:,'TOTAL COUNT']


month_analysis.plot.line(x='Month',color='red', linewidth=4.0, legend=None, fontsize=16)
plt.xlabel("")
plt.gcf().set_size_inches(14,6)
plt.ylim([0, 250])
plt.savefig('wave 1 sendouts.png', transparent=True,dpi=(300),edgecolor='w')
```
